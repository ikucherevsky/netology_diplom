```python
#функция построения скользящей средней
def SMA(data,period):
    sma=[]
    sma.append(sum(data[:period]) / period)
    for i in range(len(data) - period):
        sma.append(sma[i] + data[period + i] / period - data[i] / period)
    return sma
        
```

```python
# функция для преобразования изменений параметра в относительное (от начального значения) изменение
def deviation_value(data):
    dev_val = []
    for i in range(len(data)):
        dev_val.append(abs(data[0] - data[i]))
    return dev_val
        
```

```python
import numpy as np
```

```python
unit_number = 4
unit = fd_001_train[fd_001_train['unit_number']== unit_number]
unit.drop(columns=['unit_number','time_in_cycles','setting_1','setting_2'],inplace=True)
period = 6
for column in unit.columns:
    y = unit[column].to_numpy()
    initial = y[0]
    #y = (y / (initial * 1.02)) * 100
    x = np.linspace(0, len(y), len(y))[:, np.newaxis]
    x1 = np.linspace(period, len(y) , len(y) - period + 1)[:, np.newaxis]
    dv = deviation_value(y)
    Y = (dv/initial) *50 * 100
    avg = SMA(Y,period)
    plt.title(f'параметр {column} для {unit_number} двигателя, 1 значение {initial}, макс износ {round(Y.max())} %, +2 % {initial * 1.02}, avg max = {round(max(avg),2)}')
    if Y.max() > 100 :
        plt.plot(x, Y, c="r", linewidth=2)
    else:
        plt.plot(x, Y, c="g", linewidth=2)
    plt.plot(x1, avg, c="b", linewidth=2)
    plt.show()
```
